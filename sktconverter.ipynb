```
from indic_transliteration import sanscript
from indic_transliteration.sanscript import SchemeMap, SCHEMES, transliterate
import os
import re

scheme_map = SchemeMap(SCHEMES[sanscript.VELTHUIS], SCHEMES[sanscript.IAST])
scheme_map_san = SchemeMap(SCHEMES[sanscript.VELTHUIS], SCHEMES[sanscript.DEVANAGARI])
f=open(os.path.join('/content/drive/MyDrive/data/skt-files/chapter2-final.skt'))
print("reading file: /content/drive/MyDrive/data/skt-files/chapter2-final.skt")
data=f.read()
skttag=re.findall(r'\{\\skt\s(.*?)\}',data,re.DOTALL|re.MULTILINE)
sktitag=re.findall(r'\{\\skti\s(.*?)\}',data,re.DOTALL|re.MULTILINE)

newdata=data

for velword in skttag:
  if(newdata.find('{\skt '+velword+'}')!=-1):
    newdata=newdata.replace('{\skt '+velword+'}',transliterate(velword, scheme_map=scheme_map_san))
  #if space is not found then try newline
  elif(newdata.find('{\skt\\n'+velword+'}')!=-1):
    newdata=newdata.replace('{\skt\n'+velword+'}',transliterate(velword, scheme_map=scheme_map_san))
  #if nothing is found log warn message
  else:
    print("No replacement found for: "+"{\skt\\n"+velword+"}")


    
for velword in sktitag:
  if(newdata.find('{\skti '+velword+'}')!=-1):
    newdata=newdata.replace('{\skti '+velword+'}',transliterate(velword, scheme_map=scheme_map))
  #if space is not found then try newline
  elif(newdata.find('{\skti\\n'+velword+'}')!=-1):
    newdata=newdata.replace('{\skt\n'+velword+'}',transliterate(velword, scheme_map=scheme_map_san))
  #if nothing is found log warn message
  else:
    print("No replacement found for: "+"{\skt\\n"+velword+"}")

print('write to outputfile')
with open('/content/drive/MyDrive/data/skt-files-out/chapter2-final.tex','w') as f:
  f.write(newdata)
```
